```python
import pandas as pd
infile = open('C:/Users/18518/LIA/text analytics/project/Tweets/4-13-FindingYourFeet.csv', 'r')
```

```python
df = pd.read_csv(infile)
df.columns
```

```python
text=df['text']
text=list(text)
text
time=list(df['Time'])
favorit=list(df['favorit'])
followers=list(df['followers'])
username=list(df['username'])
retweet_count=list(df['retweet_count'])
```

```python
c=-1
review=[]
count=[]
ntime=[]
nfavorit=[]
nfollowers=[]
nusername=[]
nretweet_count=[]
while c <= len(text):
    for line in text:
        c+=1
        if line not in review:
            count.append(c)
            review.append(line)
            ntime.append(time[c])
            nfavorit.append(favorit[c])
            nfollowers.append(followers[c])
            nusername.append(username[c])
            nretweet_count.append(retweet_count[c])

final=[('text',review),('count',count),('time',ntime),('favorit',nfavorit),('followers',nfollowers),('retweet_count',nretweet_count),('username',nusername)]
df = pd.DataFrame.from_items(final)
df
```

```python
df.to_csv('4-13-FindingYourFeet.csv')
```
