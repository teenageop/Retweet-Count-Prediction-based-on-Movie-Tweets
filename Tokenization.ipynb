```python
import nltk 
from nltk.corpus import stopwords
from nltk import FreqDist
```

```python
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
```

```python
infile = open('C:/Users/18518/LIA/text analytics/project/Cleaned Tweets/ALL-ReadyPlayerOne_cleaned.csv','r',encoding='utf-8')
df = pd.read_csv(infile)
# Extract the reviews and convert them to a list
reviews=df['text']
reviews=list(reviews)
new_reviews=[]
for review in reviews:
    review = str(review).replace('\\n','')
    review=review.replace('b','')
    review=review.replace('RT','')
    review=review.replace('\\xe2\\x80\\x98s','')
    review=review.replace('\\xe2\\x80\\xa6','')
    words = review.split(' ')
    for word in words:
        if '\\' in word or '@' in word or 'https:' in word or '//' in word:
            words.remove(word)
    new_review=' '.join(words)
    new_reviews.append(new_review)
print(new_reviews)

```

```python
reviews_final=[]
for review in new_reviews:
    if review not in reviews_final:
        reviews_final.append(review)
        print(review)
```

```python
word2=[]
word_freq={}
for k in word_freq.keys():
    word_freq.setdefault(k,0)
for rev in reviews_final:
    tokens = nltk.word_tokenize(rev)
    mystopwords = stopwords.words('english')
    words = [w.lower() for w in tokens if w.isalpha() if w.lower() not in mystopwords]
    for w in words:
        word2.append(w)
        
freq = FreqDist(word2)
sorted_freq = sorted(freq.items(),key = lambda k: k[1], reverse =True)
df=pd.DataFrame(sorted_freq[:50])
df.columns = ['word','frequency']
df['film']='Outside In'
print(df)
df.to_csv('Tokens of movies2.csv',mode='a')
```
