```python
# Import the packages that will be used
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
```

```python
# Open the infile
infile = open('C:/Users/18518/LIA/text analytics/project/Cleaned Tweets/ALL/Total FindingYourFeet New.csv','r',encoding='utf-8')
```

```python
df=pd.read_csv(infile)
df.columns
```

## Remove Stopwords

```python
# Extract the reviews and convert them to a list

reviews=df['text']
reviews=list(reviews)
new_reviews=[]
for review in reviews:
    review = review.replace('\\n','')
    review=review.replace('b','')
    review=review.replace('RT','')
    review=review.replace('\\xe2\\x80\\x98s','')
    review=review.replace('\\xe2\\x80\\xa6','')
    words = review.split(' ')
    for word in words:
        if '\\' in word or '@' in word or 'https:' in word or '//' in word:
            words.remove(word)
    new_review=' '.join(words)
    new_reviews.append(new_review)
print(new_reviews)
```

## Calculate sentiment score

```python
# Use Sentiment Analyzer to analyze the sentiment score for each review
sid = SentimentIntensityAnalyzer()
ssl=[]
c=0
for review in new_reviews:
     c+=1
     nltk.sentiment.util.demo_liu_hu_lexicon(review, plot=False)
     ss = sid.polarity_scores(review)
     print(ss)
     ssl.append(ss)
```

```python
# Turn the list
df=pd.DataFrame(ssl)
df
```

```python
df.to_csv('Total FindingYourFeet New.csv')
```
